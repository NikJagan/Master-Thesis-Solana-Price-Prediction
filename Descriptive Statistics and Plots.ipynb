```python
# Plot the Class Distribution Histogram of the Solana Trend Variables

# Set the aesthetic style of the plots
sns.set_style("white")

# Create the figure and axis
plt.figure(figsize=(10, 6))

# Plot the histogram
ax = sns.histplot(data=df, x='sol_trend', bins=2, color='darkblue', edgecolor='black', shrink=0.8)

# Add labels and title with larger font sizes
ax.set_xlabel('Trend Class', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)

# Set x-axis labels to only 0 and 1 with corresponding names
ax.set_xticks([0.25, 0.75])
ax.set_xticklabels(['Downtrend', 'Uptrend'], fontsize=12)
ax.tick_params(axis='y', labelsize=12)

# Annotate the frequency on top of the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', 
                 xytext=(0, 5), textcoords='offset points', fontsize=12, color='black')

# Show the plot
plt.show()
```

```python
# Remove 'sol_trend' column from df
df_without_sol_trend = df.drop(columns=['sol_trend'])

# Compute the correlation matrix
corr_matrix = df_without_sol_trend.corr()

# Plot the heatmap
plt.figure(figsize=(16, 13))  # Adjust the size to be more compact
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, 
            cbar_kws={'shrink': .5},  # Shrink the color bar
            xticklabels=True, yticklabels=True, annot_kws={"size": 15})  # Increase the font size of the numbers

# Adjust font sizes
plt.xticks(fontsize=15)  # Reduce font size of x-axis labels
plt.yticks(fontsize=15)  # Reduce font size of y-axis labels
plt.show()
```

```python
# Filter columns that contain 'close' in their names and include 'CPILFESL'
filtered_columns = [col for col in df.columns if 'close' in col.lower()] + ['CPILFESL']

# Calculate comprehensive descriptive statistics
desc_stats = df[filtered_columns].describe().T

# Rename columns for clarity
desc_stats.rename(columns={
    'count': 'Count',
    'mean': 'Mean',
    'std': 'Standard Deviation',
    'min': 'Minimum',
    '25%': '1st Quartile',
    '50%': 'Median',
    '75%': '3rd Quartile',
    'max': 'Maximum'
}, inplace=True)

# Display the descriptive statistics
print(desc_stats)
```
